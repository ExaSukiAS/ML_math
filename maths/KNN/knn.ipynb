```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
df
```

```python
result_varience = 2
```

```python
target = np.array([20.5, 7])
```

```python
def get_distance(x, y):
    return np.sqrt(np.sum((x - y) ** 2))
```

```python
df["Distance"] = df.apply(
    lambda row: get_distance(row[["Hours_of_Study", "Hours_of_Sleep"]], target), axis=1
)
df
```

```python
for i in range(0, len(df)):
    for j in range(0, len(df) - (i+1)):
        if df.loc[j][3] > df.loc[j + 1][3]:
            df.loc[j], df.loc[j + 1] = df.loc[j + 1], df.loc[j]
df
```

```python
k = int(np.sqrt(len(df)))
if k % 2 == 0:
    k += 1
k
```

```python
results  = np.zeros(result_varience, dtype=int)
for i in range (0, k-1):
    results[df.Result[k]] += 1
results
```

```python
prev_num = 0
highest_num = 0
highest_index = 0
for i in range(0, len(results)-1):
    prev_num = results[i]
    curr_num = results[i+1]
    if curr_num > prev_num:
        highest_num = curr_num
        highest_index = i+1
print(f"{target} goes to {highest_index} class")
```
